3 Image Compression and Decompression using PCA.
Load and Prepare Data:
â€¢ Fetch an image of you choice.{If colour convert to grayscale}

â€¢ Center the dataset - Standaridze the Data.

â€¢ Calculate the covaraince matrix of the Standaridze data.

```python
# Convert to NumPy array
image_array = np.array(img)
# Get image dimensions
height, width = image_array.shape
# Reshape the image into a 2D array where each row represents a row of pixels
# This allows PCA to process each row as an observation and find patterns across pixel intensities
data = image_array.reshape( height , width )
```

```python
# Compute the mean of each column ( feature )
mean = np.mean( data , axis =0)
# Subtract mean to center the data
centered_data = data - mean

# Compute covariance matrix
cov_matrix = np.cov( centered_data , rowvar = False )
print(cov_matrix)
```

3. Eigen Decomposition and Identifying Principal Components:
â€¢ Compute Eigen Values and Eigen Vectors.

â€¢ Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues.

â€¢ Identify the Principal Components with the help of cumulative Sum plot.


```python
# Compute eigenvalues and eigenvectors
eigenvalues , eigenvectors = np.linalg.eigh( cov_matrix )
# Sort eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np . argsort ( eigenvalues ) [:: -1]
eigenvalues = eigenvalues [ sorted_indices ]
eigenvectors = eigenvectors [: , sorted_indices]
```

```python
# Plot the explained variance ratio
explained_variance_ratio = eigenvalues / np .sum ( eigenvalues )
plt . plot ( np . cumsum ( explained_variance_ratio ) )
plt . title (" Cumulative Explained Variance ")
plt . xlabel (" Number of Components ")
plt . ylabel (" Cumulative Explained Variance ")
plt . grid ( True )
plt . show ()
```

```python
k = 20 # Choose k principal components
components = eigenvectors [: , : k ]
```

Reconstruction and Experiment:

â€¢ Reconstruction: Transform the original data by multiplying it with the selected eigenvec- tors(PCs) to obtain a lower-dimensional representation.


â€¢ Experiments: Pick Four different combination of principal components with various ex- plained variance value and compare the result.


â€¢ Display the Results and Evaluate.

```python
# Project the data onto the principal components
compressed_data = np.dot(centered_data, components)
print(compressed_data.shape)

# Reconstruct the image from compressed data
decompressed_data = np.dot( compressed_data , components . T ) + mean
```

```python
plt.figure ( figsize =(12 , 6) )

# Original Image
plt.subplot (1 , 3 , 1)
plt . imshow ( img, cmap="gray" )
plt . title (" Original Image ")

# Compressed Representation
plt . subplot (1 , 3 , 2)
plt . imshow ( compressed_data,  cmap="gray")
plt . title ( f" Compressed Image (k={k})")
# Decompressed Image
plt . subplot (1 , 3 , 3)
plt . imshow ( decompressed_data,  cmap="gray")
plt . title (" Decompressed Image ")
plt . tight_layout ()
plt . show ()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def compress_and_reconstruct(image_array, k):
    """Compresses and reconstructs an image using PCA with k components."""
    height, width = image_array.shape
    data = image_array.reshape(height, width)
    mean = np.mean(data, axis=0)
    centered_data = data - mean
    cov_matrix = np.cov(centered_data, rowvar=False)
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]
    components = eigenvectors[:, :k]
    compressed_data = np.dot(centered_data, components)
    decompressed_data = np.dot(compressed_data, components.T) + mean
    return decompressed_data

image_array = np.array(img)
k_values = [20, 50, 100, 150]  # Different values of k

# Create subplots for each k value
fig, axes = plt.subplots(1, len(k_values), figsize=(15, 5))

for i, k in enumerate(k_values):
    reconstructed_image = compress_and_reconstruct(image_array, k)
    axes[i].imshow(reconstructed_image, cmap='gray')
    axes[i].set_title(f'k = {k}')
    axes[i].axis('off')

plt.tight_layout()
fig.suptitle("Comparison of different values of k on the image", y=0.05)
plt.show()
```

**# Conclusion: Impact of Different k Values in PCA**

In Principal Component Analysis (PCA), the number of principal components, k, determines the level of image compression and the extent to which image quality is retained. Based on the observations from the plot, we can summarize the effects of different k values as follows:

k = 20

The reconstructed image appears significantly blurred.
Fine details and textures are largely lost.
This represents a high level of compression but at the cost of poor visual quality.

k = 50

Image quality improves compared to k = 20.
Some edges and features become visible, though finer details remain missing.
This provides a moderate compression level with reasonable quality.

k = 100

The reconstructed image closely resembles the original.
Most essential details and textures are retained.
This strikes a good balance between compression and visual clarity.

k = 150

The image is nearly identical to the original.
Minimal loss in quality is observed.
This represents low compression but ensures high reconstruction accuracy.

Key Insights**

A smaller k leads to higher compression but results in lower image quality.
A larger k retains more visual details but sacrifices compression efficiency.
The optimal k value depends on the specific applicationâ€”whether the priority is saving storage space or maintaining visual fidelity.
For the given image, k = 100 appears to preserve most of the quality while still achieving compression.
This highlights the fundamental trade-off in PCA-based image compression between data reduction and reconstruction accuracy.
